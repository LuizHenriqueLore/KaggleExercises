```python
iris_path = '/kaggle/input/iris/Iris.csv'
iris_data = pd.read_csv(iris_path)
print(iris_data.shape)
iris_data.head()
```

```python
from sklearn.model_selection import train_test_split
#Target
y = iris_data.Species

#Features
features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
X = iris_data[features]

#Split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

#CreateModel
iris_random_forest = RandomForestClassifier(n_estimators=100, random_state=0)

#Fit training data
iris_random_forest.fit(X_train, y_train)
```

```python
# Generate test predictions
predictions_valid = iris_random_forest.predict(X_valid)

# Save predictions in format used for competition scoring
#output = pd.DataFrame({'Id': X_test.index,
#                       'SalePrice': preds_test})
#output.to_csv('submission.csv', index=False)

#print(predictions_valid)
#print(y_valid)
index = 0
predictions_err = 0
y_list = y_valid.values.tolist()
for predict in predictions_valid:
    if y_list[index] != predict :
       predictions_err += 1
    index += 1
    
print('errors: ',predictions_err)
```
